```python
import os
from glob import glob #filename pattern matching module
import xlrd
import openpyxl 
import datetime as dt
from datetime import datetime
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
```

```python
cwd = os.getcwd()
```

```python
# the day of analysis
today = str(np.datetime64('2023-11-05'))
past_7days = str(np.datetime64(today)-np.timedelta64(27,'D'))
```

```python
date_range = pd.date_range(past_7days, today, freq = "1D")
print(date_range)
```

```python
date_range = [date_range.strftime('%Y-%m-%d') for date_range in date_range]
date_range # to str
```

```python
#for 0725 all_channel_data
namelist=[]
for name in date_range:
    files_path = glob("C:\\Users\\buche\\Desktop\\travail_workspace\\ETL_workspace\\all_channel_rawdata\\" + name + "\\post_deduction\\channel_*.xlsx")
    # files_path = glob("D:\\Desktop\\travail_workspace\\ETL_workspace\\all_channel_rawdata\\" + name + "\\post_deduction\\channel_*.xlsx")
    namelist.append(files_path)
    print(namelist)
```

```python
namelist = [i for name in namelist for i in name]
```

```python
#合併檔案
all_ch = pd.DataFrame()
for file in namelist:
    df = pd.read_excel(file, engine='openpyxl')
    all_ch = all_ch.append(df, ignore_index=True)
```

```python
# 刪除7002-7021，'官网渠道', '代理邀请'
def splitname(x):
    return x[0:4]
l1 = all_ch['渠道名称'].tolist()
l1 = list(map(splitname, l1))
all_ch['渠道名称']=l1
```

```python
excluded_ch = list(np.array(range(7002,7021), dtype=str))
official_ch = ['官网渠道', '代理邀请']
del_ch = excluded_ch + official_ch
del_ch = all_ch['渠道名称'].isin(pd.Series(del_ch))
```

```python
all_ch = all_ch[-del_ch]
```

```python
# 重整
all_ch = all_ch.sort_values(by='渠道名称')
all_ch.reset_index(drop=True, inplace=True)
```

```python
all_ch['渠道名称']=all_ch['渠道名称']+'渠道'
```

```python
all_ch = pd.DataFrame()
for file in files:
    df = pd.read_excel(file, engine='openpyxl')
    all_ch = all_ch.append(df)
```

```python
# 製作第一組圖表:充值人數、新增註冊用戶數、轉換率（新增）
all_ch['转换率']= round((all_ch['充值人数']/all_ch['新增注册用户数'])*100, 2)
# nan, inf -> 0
all_ch['转换率'] = all_ch['转换率'].fillna(0)
all_ch.loc[(all_ch['转换率']==np.inf), '转换率'] = 0

# 轉百分比
#all_ch['转换率']= str((all_ch['充值人数']/all_ch['新增注册用户数'])*100)[:5] + '%'
#all_ch['转换率']= "%.2f%%" % ((all_ch['充值人数']/all_ch['新增注册用户数'])*100)
#all_ch['转换率'] = all_ch['转换率'].apply(lambda x: format(x, '.2%'))
```

```python
# 篩選需要欄位
df_convert = all_ch.iloc[:, [0, 2, 5, 14, 42]]
```

```python
# 篩選需要時間
today = str(np.datetime64('2023-07-18') )# 之後改回來today = str(np.datetime64('today'))
past_week = str(np.datetime64(today)-np.timedelta64(30,'D')) # 月報拉30天資料
filt = (df_convert['统计日期'].between(past_week, today)) # 日報需前六天
#filt
df_convert = df_convert.loc[filt]
```

```python
# 要從7021開始做，將渠道名分割留下數字部分
def splitname(x):
    return x[0:4]
l1 = df_convert['渠道名称'].tolist()
l1 = list(map(splitname, l1))
df_convert['渠道名称']=l1
```

```python
import os
from glob import glob #filename pattern matching module
import xlrd
import openpyxl 
import datetime as dt
from datetime import datetime
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
```

```python
today = str(np.datetime64('2023-10-31')) 
print(today)
past_4month = str(np.datetime64(today)-np.timedelta64(122,'D'))
print(past_4month)
```

```python
date_range = pd.date_range(past_4month, today, freq = "1D")
date_range = [date_range.strftime('%Y-%m-%d') for date_range in date_range]
print(date_range)
print(len(date_range))
```

## 渠道總表

```python
#for 0725 all_channel_data
namelist=[]
for name in date_range:
    files_path = glob("C:\\Users\\buche\\Desktop\\travail_workspace\\ETL_workspace\\all_channel_rawdata\\" + name + "\\pre_deduction\\channel_*.xlsx")
    namelist.append(files_path)
    print(namelist)
```

```python
namelist = [i for name in namelist for i in name]
```

```python
all_ch = pd.DataFrame()
for file in namelist:
    df = pd.read_excel(file, engine='openpyxl')
    all_ch = all_ch.append(df, ignore_index=True)
```

```python
all_ch[all_ch.duplicated()]
```

```python
all_ch.to_csv('C:\\Users\\buche\\Desktop\\travail_workspace\\ch_report\\M10_ch_'+ today +'.csv', encoding = 'utf-8-sig') # 月報用
```

## 篩選製圖df

```python
a = all_ch
```

```python
# 新增轉換率
a['转换率'] = round(a['充值人数']/a['新增注册用户数'], 4)
# nan, inf -> 0
a['转换率'] = a['转换率'].fillna(0)
a.loc[(a['转换率']==np.inf), '转换率'] = 0
a['转换率']= a['转换率'].apply(lambda x: format(x, '.2%'))
```

```python
# 選擇需要欄位
a = a.iloc[:, [0, 2, 4, 5, 7, 34]]
a
```

```python
# 要從7021開始做，將渠道名分割留下數字部分
def splitname(x):
    return x[0:4]
l1 = a['渠道名称'].tolist()
l1 = list(map(splitname, l1))
a['渠道名称']=l1
```

```python
# 刪除7002-7021，'官网渠道', '代理邀请'
excluded_ch = list(np.array(range(7002,7021), dtype=str))
official_ch = ['官网渠道', '代理邀请']
del_ch = excluded_ch + official_ch
del_ch = a['渠道名称'].isin(pd.Series(del_ch))
a = a[-del_ch]
a['渠道名称']=a['渠道名称']+'渠道'
a = a.sort_values(by=['统计日期','渠道名称'], ascending=True).reset_index(drop = True)
a
```

```python
a.to_csv('C:\\Users\\buche\\Desktop\\travail_workspace\\ch_report\\M10_filte_ch_'+ today +'.csv', encoding = 'utf-8-sig') # 月報用
```

## 扣後

```python
namelist=[]
for name in date_range:
    files_path = glob("C:\\Users\\buche\\Desktop\\travail_workspace\\ETL_workspace\\all_channel_rawdata\\" + name + "\\post_deduction\\channel_*.xlsx")
    # files_path = glob("D:\\user\\Desktop\\travail_workspace\\ETL_workspace\\all_channel_rawdata\\" + name + "\\pre_deduction\\channel_*.xlsx")
    namelist.append(files_path)
    print(namelist)
namelist = [i for name in namelist for i in name]
#合併檔案
post_deduction = pd.DataFrame()
for file in namelist:
    df = pd.read_excel(file, engine='openpyxl')
    post_deduction = post_deduction.append(df, ignore_index=True)
```

```python
# 要從7021開始做，將渠道名分割留下數字部分
def splitname(x):
    return x[0:4]
l1 = post_deduction['渠道名称'].tolist()
l1 = list(map(splitname, l1))
post_deduction['渠道名称']=l1

# 刪除7002-7020，'官网渠道', '代理邀请'
excluded_ch = list(np.array(range(7002,7021), dtype=str))
official_ch = ['官网渠道', '代理邀请']
del_ch = excluded_ch + official_ch
del_ch = post_deduction['渠道名称'].isin(pd.Series(del_ch))

post_deduction = post_deduction[-del_ch]
```

```python
post_deduction['渠道名称']=post_deduction['渠道名称']+'渠道'
```

## create pivot tb 

```python
el_daterange = pd.DataFrame(date_range)
```

```python
el_daterange = el_daterange.replace('2023-', '', regex=True)
```

```python
el_daterange
```

```python
# 8date list
li_week = np.array_split(np.array(el_daterange).flatten(), 2)
li_week

#?  1
#? 30/31?
```

```python
rank = list(range(0, 2))
```

```python
rank 
```

```python
df_M = pd.DataFrame()
for i in rank:
    per_M = t[t[''].isin(li_week[i])]
    per_M['week_rank'] = i+1
    print(per_M)
```

```python
df_week = pd.DataFrame()
for i in rank:
    per_week = t[t[''].isin(li_week[i])]
    per_week['week_rank'] = i+1
    per_week['2'] = per_week['']
    per_week_pivot = pd.pivot_table(per_week, index=["", "week_rank"], aggfunc={"": np.sum , "2": 'count', "id": pd.Series.nunique})
    per_week_pivot['mean_order'] = round(per_week_pivot['']/per_week_pivot['2'], 2)
    per_week_pivot['mean_person'] = round(per_week_pivot['']/per_week_pivot['id'], 2)

    df_week = df_week.append(per_week_pivot)
```

```python
df_week = df_week.sort_values(by=["", "week_rank"])
```

```python
df_week['amount_dod'] = pd.Series(df_week['']).pct_change().apply(lambda x: format(x, '.2%')) # 
df_week['order_dod'] = pd.Series(df_week['2']).pct_change().apply(lambda x: format(x, '.2%')) # 
df_week['person_dod'] = pd.Series(df_week['id']).pct_change().apply(lambda x: format(x, '.2%')) # 
df_week['mean_order_dod'] = pd.Series(df_week['mean_order']).pct_change().apply(lambda x: format(x, '.2%')) # 
df_week['mean_person_dod'] = pd.Series(df_week['mean_person']).pct_change().apply(lambda x: format(x, '.2%')) # 
```

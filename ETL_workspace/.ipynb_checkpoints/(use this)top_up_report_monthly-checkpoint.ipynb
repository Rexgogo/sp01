## create pivot tb 

```python
el_daterange = pd.DataFrame(date_range)
```

```python
el_daterange = el_daterange.replace('2023-', '', regex=True)
```

```python
el_daterange
```

```python
# 區分成8週的date list
li_week = np.array_split(np.array(el_daterange).flatten(), 2)
li_week

#? 確認 後臺資料不會有跨日狀況，否則每組資料要多拉隔月1日的資料
#? 重寫：　根據月份自動判斷隔出30/31天?
```

```python
rank = list(range(0, 2))
```

```python
rank 
```

```python
df_M = pd.DataFrame()
for i in rank:
    per_M = t[t['日期'].isin(li_week[i])]
    per_M['week_rank'] = i+1
    print(per_M)
```

```python
df_week = pd.DataFrame()
for i in rank:
    per_week = t[t['日期'].isin(li_week[i])]
    per_week['week_rank'] = i+1
    per_week['充值金额2'] = per_week['充值金额']
    per_week_pivot = pd.pivot_table(per_week, index=["支付类型", "week_rank"], aggfunc={"充值金额": np.sum , "充值金额2": 'count', "用户id": pd.Series.nunique})
    per_week_pivot['mean_order'] = round(per_week_pivot['充值金额']/per_week_pivot['充值金额2'], 2)
    per_week_pivot['mean_person'] = round(per_week_pivot['充值金额']/per_week_pivot['用户id'], 2)

    df_week = df_week.append(per_week_pivot)
```

```python
df_week = df_week.sort_values(by=["支付类型", "week_rank"])
```

```python
df_week['amount_dod'] = pd.Series(df_week['充值金额']).pct_change().apply(lambda x: format(x, '.2%')) # 區分支付類型
df_week['order_dod'] = pd.Series(df_week['充值金额2']).pct_change().apply(lambda x: format(x, '.2%')) # 區分支付類型
df_week['person_dod'] = pd.Series(df_week['用户id']).pct_change().apply(lambda x: format(x, '.2%')) # 區分支付類型
df_week['mean_order_dod'] = pd.Series(df_week['mean_order']).pct_change().apply(lambda x: format(x, '.2%')) # 區分支付類型
df_week['mean_person_dod'] = pd.Series(df_week['mean_person']).pct_change().apply(lambda x: format(x, '.2%')) # 區分支付類型
```

```python
for i in date_range:
    print(type(i))
    
# pre-deduction
    files_path = glob("C:\\Users\\buche\\Desktop\\travail_workspace\\ETL_workspace\\all_channel_rawdata\\" + i.strftime('%Y-%m-%d') + "\\pre_deduction\\channel_*.xlsx")
    # files_path = glob("D:\\Desktop\\travail_workspace\\ETL_workspace\\all_channel_rawdata\\" + date_str.all() + "\\pre_deduction\\channel_*.xlsx")
    
    pre_deduction = pd.DataFrame()
    for file in files_path:
        df = pd.read_excel(file, engine='openpyxl')
        pre_deduction = pre_deduction.append(df, ignore_index=True).sort_values(by=['渠道名称'], ignore_index=True)
    pre = pre_deduction
    
# post-deduction
    files_path = glob("C:\\Users\\buche\\Desktop\\travail_workspace\\ETL_workspace\\all_channel_rawdata\\" + i.strftime('%Y-%m-%d') + "\\post_deduction\\channel_*.xlsx")
    # files_path = glob("D:\\Desktop\\travail_workspace\\ETL_workspace\\all_channel_rawdata\\" + date_str.all() + "\\post_deduction\\channel_*.xlsx")

    post_deduction = pd.DataFrame()
    for file in files_path:
        df = pd.read_excel(file, engine='openpyxl')
        post_deduction = post_deduction.append(df, ignore_index=True).sort_values(by=['渠道名称'], ignore_index=True)
    post = post_deduction

# google_all_ch: 取當月個別渠道(如果日期為月1號，要參照"當月-1"的表) 
    c = str(i.month - 1) + '月个别渠道分析（官网）'
    b = str(i.month) + '月个别渠道分析（官网）'
    if i.day==1:
        google_all_ch = completed_data.worksheet_by_title(c).get_as_df()
    else:
        google_all_ch = completed_data.worksheet_by_title(b).get_as_df()
    
    google_all_ch = google_all_ch.rename(columns={'新增注册用户数': 'as昨日安装'})
    
    # google_final_sheet: 取當月總表(如果日期為當月1號，要參照"當月-1"的總表)
    g = str(i.month - 1) + '月总表'
    h = str(i.month) + '月总表'
    if i.day==1:
         google_final_sheet = completed_data.worksheet_by_title(g).get_as_df()
    else:
         google_final_sheet = completed_data.worksheet_by_title(h).get_as_df() 
    
    # google_dod_sheet:取當月7月总数据(如果日期為月1號，要參照"當月-1"的7月总数据)
    k = str(i.month - 1) + '月总数据'
    m = str(i.month) + '月总数据'
    if i.day==1:
         google_dod_sheet = completed_data.worksheet_by_title(k).get_as_df()
    else:
         google_dod_sheet = completed_data.worksheet_by_title(m).get_as_df()
    
    
# 列出 pre 活躍用戶不等於0 的渠道名稱
    filt = pre['活跃用户数']!=0
    namelist = pre.loc[filt, '渠道名称'].values.tolist()
# 排除 活躍用户=0 的渠道
    pre = pre[pre['活跃用户数']!=0]
# 根據 pre有的渠道名稱篩出 post表
    df = pd.DataFrame()
    for j in namelist:
        a = post[post['渠道名称'].apply(lambda x: x == j)]
        df = df.append(a, ignore_index=True)

    # g的新增注册用户数 -> pre的扣量後註冊
    pre.insert(5, column='扣量后注册', value=df['新增注册用户数'].values) 
    ch_sheet = pre.reset_index()
    
    ch_sheet.insert(4, column='单价', value=2)
    # 根據渠道名填入扣量比例
    ch_sheet = ch_sheet.merge(de_percentage[['渠道名称', '百分比']], how='left', on='渠道名称').rename(columns={'百分比': '扣量％数'})
    
    ch_sheet.set_index("渠道名称", inplace=True)
    ch_sheet.at["代理邀请","扣量后注册"]=0
    ch_sheet.at["官网渠道","扣量后注册"]=0
    ch_sheet = ch_sheet.reset_index()
    
    yest_ch = google_all_ch[google_all_ch['统计日期'] == yest_data]
    
    
    ch_sheet = ch_sheet.merge(yest_ch[['渠道名称', 'as昨日安装']], how='left', on='渠道名称')
    ch_sheet = ch_sheet.rename(columns={'as昨日安装':'昨日安装'})
    

# 更改["昨日安装"]屬性, nan, inf -> 0

    if ch_sheet["昨日安装"].isna().sum() > 0:
        ch_sheet["昨日安装"] = ch_sheet["昨日安装"].fillna(0)

    ch_sheet["昨日安装"] = ch_sheet["昨日安装"].astype(int)
    
    ch_sheet["对比昨日"] = ch_sheet["新增注册用户数"]-ch_sheet["昨日安装"]
    ch_sheet["对比昨日"] = ch_sheet["对比昨日"].fillna(0)
    
    ch_sheet['昨日同比率'] = ch_sheet['新增注册用户数'] / ch_sheet['昨日安装']

    if (ch_sheet['昨日同比率']==np.inf).sum() > 0:
        ch_sheet.loc[(ch_sheet['昨日同比率']==np.inf), '昨日同比率'] = 0
    if ch_sheet["昨日同比率"].isna().sum() > 0:
        ch_sheet['昨日同比率'] = ch_sheet['昨日同比率'].fillna(0)
        
# 調整百分比格式->str
    ch_sheet['昨日同比率'] = ch_sheet['昨日同比率'].apply(lambda x: format(x, '.0%'))
    
    ch_sheet['扣量前费用'] = ch_sheet['单价']* ch_sheet['新增注册用户数']
    ch_sheet['扣量后费用'] = ch_sheet['单价']* ch_sheet['扣量后注册']
   
    ch_sheet.set_index("渠道名称", inplace=True)
    ch_sheet.at["代理邀请","扣量前费用"]=0
    ch_sheet.at["官网渠道","扣量前费用"]=0
    ch_sheet.at["代理邀请","扣量后费用"]=0
    ch_sheet.at["官网渠道","扣量后费用"]=0
    ch_sheet = ch_sheet.reset_index()
    
    ch_sheet["提现金额"] = ch_sheet["提现金额"].fillna(0)
    ch_sheet["提现金额"] = ch_sheet["提现金额"].astype(int)
    
    ch_sheet['渠道营收'] = ch_sheet['总充值金额']-ch_sheet['提现金额']-ch_sheet['扣量后费用']
    
    amount_deduction = (ch_sheet['新增注册用户数'] - ch_sheet['扣量后注册']).to_frame()
    
# sheet_A = 總表
    ch_record1_ytd = google_final_sheet[google_final_sheet['统计日期']==yest_data] 
    
    # 建空df存放今日數據
    use_col = ch_record1_ytd.columns.tolist()
    ch_record1_td = pd.DataFrame(columns=use_col, index = range(1, 2))
    
    # 統計日期
    ch_record1_td.iloc[0, 0]= i.strftime('%Y-%m-%d')
    
    # import每日統計表，dict轉成df 
    daily_stats = pd.read_excel('C:\\Users\\buche\\Desktop\\travail_workspace\\ETL_workspace\\all_channel_rawdata\\' + pd.to_datetime(i).strftime('%Y-%m-%d') + '\\daily_stats.xlsx')#, sheet_name = [0, '每日统计数据'])
    # daily_stats = pd.read_excel('D:\\Desktop\\travail_workspace\\ETL_workspace\\all_channel_rawdata\\' + date_str.all() + '\\daily_stats.xlsx')
    
    yest_data = pd.to_datetime(i-np.timedelta64(1,'D')).strftime('%Y-%m-%d')
    yest_daily_stats= pd.read_excel('C:\\Users\\buche\\Desktop\\travail_workspace\\ETL_workspace\\all_channel_rawdata\\' + yest_data + '\\daily_stats.xlsx')
    
    # 登录用户数
    ch_record1_td.iloc[0,1] = daily_stats.iloc[0, 22]

    # 登录同比 還沒轉百分比
    ch_record1_td.iloc[0,2] = daily_stats.iloc[0, 22] / yest_daily_stats.iloc[0, 22]

    # 新增注册用户数
    ch_record1_td.iloc[0,3] = daily_stats.iloc[0, 28]

    # 新增注册同比 還沒轉百分比
    ch_record1_td.iloc[0,4] = daily_stats.iloc[0, 28] / yest_daily_stats.iloc[0, 28]

    # 扣量后注册
    ch_record1_td.iloc[0,5] = ch_sheet['扣量后注册'].sum()
    # 留存用户
    ch_record1_td.iloc[0,6] = ch_record1_td.iloc[0,1]-ch_record1_td.iloc[0,3]
    # 留存同比
    ch_record1_td.iloc[0,7] = ch_record1_td.iloc[0,6] / (yest_daily_stats.iloc[0, 22]-yest_daily_stats.iloc[0, 28])

    # 昨日安装
    ch_record1_td.iloc[0,8] = yest_daily_stats.iloc[0, 28]
    # 对比昨日
    ch_record1_td.iloc[0,9] = ch_record1_td.iloc[0,3] - yest_daily_stats.iloc[0, 28]
    # 昨日同比率 還沒轉百分比
    ch_record1_td.iloc[0,10] = ch_record1_td.iloc[0,3] / yest_daily_stats.iloc[0, 28]

    # 扣量前费用 
    ch_record1_td.iloc[0,11] = ch_sheet['扣量前费用'].sum()

    # 扣量后费用
    ch_record1_td.iloc[0,12] = ch_sheet['扣量后费用'].sum()

    # 提现（元）
    ch_record1_td.iloc[0,14] = daily_stats.iloc[0, 58]

    # 充值人数
    ch_record1_td.iloc[0,15] = daily_stats.iloc[0, 34]
    # 充值人数同比 還沒轉百分比
    ch_record1_td.iloc[0,16] = daily_stats.iloc[0, 34] / yest_daily_stats.iloc[0, 34]


    # 新用户充值
    ch_record1_td.iloc[0,17] = daily_stats.iloc[0, 52] 
    # 新用户充值人数
    ch_record1_td.iloc[0,18] = daily_stats.iloc[0, 35] 

    # 新增会员数
    ch_record1_td.iloc[0,19] =  daily_stats.iloc[0, 40] 
    # 新增会员数同比
    ch_record1_td.iloc[0,20] = daily_stats.iloc[0, 40] / yest_daily_stats.iloc[0, 40] 

    # 总充值金额
    ch_record1_td.iloc[0,21] = daily_stats.iloc[0, 46]

    # 渠道营收
    ch_record1_td.iloc[0,13] = daily_stats.iloc[0, 46]-daily_stats.iloc[0, 58]-ch_sheet['扣量后费用'].sum()

    namelist=[]
    files_path = glob("C:\\Users\\buche\\Desktop\\travail_workspace\\ETL_workspace\\all_channel_rawdata\\" + pd.to_datetime(i).strftime('%Y-%m-%d') + "\\top_up\\charge_*.xlsx")
    # files_path = glob("D:\\Desktop\\travail_workspace\\ETL_workspace\\all_channel_rawdata\\" + date_str.all() + "\\top_up\\charge_*.xlsx")
    
    namelist.append(files_path)
    namelist = [i for name in namelist for i in name]
    
    #合併檔案
    all_top_up = pd.DataFrame()
    for file in namelist:
        df = pd.read_excel(file, engine='openpyxl')
        all_top_up = all_top_up.append(df, ignore_index=True)

    t = all_top_up.iloc[:, [0, 1, 2, 3, 9, 11]]
    t = t.reset_index(drop=True)
    t.drop_duplicates(inplace=True)
    t['充值金额'] = t['充值金额'].astype(int)

    # 以['订单完成时间']為基準
    def splitname(x):
        x=str(x)
        return x[0:10]
    li_tmp = pd.Series(t["订单完成时间"])
    li_tmp = list(map(splitname,list(li_tmp)))

    t["订单完成时间2"] = li_tmp
    t["订单完成时间2"]= pd.to_datetime(t["订单完成时间2"])
    t["订单完成时间2"] = t["订单完成时间2"].apply(lambda x: x.strftime('%m-%d'))
    t.rename(columns = {"订单完成时间2":"日期"}, inplace=True)
    t['用户id'] = t['用户id'].astype(str)
    t['充值金额2'] = t['充值金额']

    pivot_t = pd.pivot_table(t, index=["支付类型", "日期"], aggfunc={"充值金额": np.sum , "充值金额2": 'count', "用户id": pd.Series.nunique})
    #　margins=True

    # set支付名稱 as index
    pivot_t.reset_index(inplace=True)
    pivot_t.set_index('支付类型', inplace=True)

    # 本日出現的支付名稱
    topup_cate = pivot_t.index

    # 所有支付名稱
    all_topup = pd.Series(['人工支付', '支付宝支付', '银行卡支付', '微信支付', '云闪付', 'QQ钱包'],name='支付类型') 

    for p in all_topup:
        if topup_cate.isin([p]).sum() != 1:
            ch_record1_td.loc[1,p] = 0
        else:
            ch_record1_td.loc[1,p] = pivot_t.loc[p, '充值金额']

    # 直推充值金额
    ch_record1_td.iloc[0,28] = daily_stats.iloc[0, 50] 

    daily_stats_suppl = key.open_by_url('https://docs.google.com/spreadsheets/d/1cLuHLzEcKfzqnio0wfPLwn-LM0KlA3UDwFteU-Bt3Bw/edit#gid=0')[0].get_as_df()

    # （旧版）苹果充值
    ch_record1_td.iloc[0,29] = daily_stats.iloc[0, 48] 
    # （旧版）苹果人均充值
    ch_record1_td.iloc[0,30] = round(daily_stats.iloc[0, 48]/daily_stats_suppl.iloc[0, 1], 2)


    # 安卓充值
    ch_record1_td.iloc[0,31] = daily_stats.iloc[0, 47] 
    # 安卓人均充值
    ch_record1_td.iloc[0,32] = round(daily_stats.iloc[0, 47]/daily_stats_suppl.iloc[0, 2], 2)

    # H5充值
    ch_record1_td.iloc[0,33] = daily_stats.iloc[0, 49] 

    # H5人均充值
    if daily_stats_suppl.iloc[0, 3] == 0:
        ch_record1_td.iloc[0,34] = 0
    else:
        ch_record1_td.iloc[0,34] = 'Error'

    # 实际拉起订单数
    ch_record1_td.iloc[0,35] = daily_stats.iloc[0, 70]

    # 拉起成功订单数
    ch_record1_td.iloc[0,36] = daily_stats_suppl.iloc[0, 4]

    # 成功订单数
    ch_record1_td.iloc[0,37] = daily_stats.iloc[0, 71]

    # 成功率
    ch_record1_td.iloc[0,38] = daily_stats.iloc[0, 71]/daily_stats.iloc[0, 70]
    # 還沒轉百分比

    # 游戏人均付费
    ch_record1_td.iloc[0,39] = (daily_stats.iloc[0, 46]-daily_stats.iloc[0, 59])/daily_stats.iloc[0, 22]
    # 還沒轉百分比


    # 游戏营收
    ch_record1_td.iloc[0,40] = daily_stats.iloc[0, 46]-daily_stats.iloc[0, 58]-ch_sheet['扣量后费用'].sum()-daily_stats.iloc[0, 59]

    # 会员总收益
    ch_record1_td.iloc[0,41] = daily_stats.iloc[0, 59]

    # 次日留存
    ch_record1_td.iloc[0,42] = daily_stats.iloc[0, 0]
    # 三日留存
    ch_record1_td.iloc[0,43] = daily_stats.iloc[0, 1]
    # 7日留存
    ch_record1_td.iloc[0,44] = daily_stats.iloc[0, 2]
    # 二次付费率
    ch_record1_td.iloc[0,45] = daily_stats.iloc[0, 6]
    # arpu
    ch_record1_td.iloc[0,46] = daily_stats.iloc[0, 7]
    # arppu
    ch_record1_td.iloc[0,47] = daily_stats.iloc[0, 13]

    sheet_A = ch_record1_td.copy(deep=False)
    
    # 填入總表
    # find the index of last row
    to_fill_excel_index = 'A' + str(i.day+1)
    print(to_fill_excel_index)
    
    # 選擇工作表
    sheet_name_for_google_final_sheet = (pd.Series(str(i.month))+ '月总表')[0]
    print(sheet_name_for_google_final_sheet)
    google_final_sheet = completed_data.worksheet_by_title(sheet_name_for_google_final_sheet)
    
    # 填入
    google_final_sheet.set_dataframe(sheet_A, to_fill_excel_index, copy_index=False, copy_head=False, extend=True, fit=False, escape_formulae=False)
    
    #________
    
    # sheet_B
    sheet_B = ch_sheet[['统计日期', '渠道ID', '渠道名称', '单价', '活跃用户数', '新增注册用户数', '扣量后注册', '扣量％数', '昨日安装', '对比昨日', '昨日同比率', '扣量前费用', '扣量后费用', '渠道营收', '充值人数', '新增会员数', '总充值金额', '直推充值金额', '提现金额', '购买会员金额', '安卓购买会员金额', 'ios购买会员金额', 'h5购买会员金额', '直推购买会员金额', '邀请购买会员金额', '内容付费金额', '打赏收入', '游戏消耗金额', '邀请返利', '次日留存', '三日留存', '7日留存', 'cpa单价', 'cpa收益', 'cps占比', 'cps收益', '毛利率', '日活付费率', '新增付费率', '二次付费率', 'arpu', 'arppu']]
    
    # 填入當日個別渠道表
    # find the index of last row
    to_fill_excel_index = 'A' + str(google_all_ch.index[-1]+3)
    print(to_fill_excel_index)
    
    # 選擇工作表
    sheet_name_for_google_all_ch = (pd.Series(str(i.month))+ '月个别渠道分析（官网）')[0]
    print(sheet_name_for_google_all_ch)
    google_all_ch = completed_data.worksheet_by_title(sheet_name_for_google_all_ch)

    # 填入
    google_all_ch.set_dataframe(sheet_B, to_fill_excel_index, copy_index=False, copy_head=False, extend=True, fit=False, escape_formulae=False)

print(i)
```

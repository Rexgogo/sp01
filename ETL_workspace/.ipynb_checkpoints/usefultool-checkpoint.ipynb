```python
files_path = glob('F:/travail_workspace/ETL_workspace/all_channel_7/all_channel*')# 把路逕存進list
print(files_path)
```

```python
#合併檔案
all_ch = pd.DataFrame()
for file in files_path:
    df = pd.read_excel(file, engine='openpyxl')
    print(file)
    all_ch = all_ch.append(df, ignore_index=True)
```

```python
col = ['统计日期', '渠道ID', '渠道名称', '新增注册用户数', '充值人数']
all_ch = all_ch[col]
all_ch
```

```python
cols = ['新增注册用户数', '充值人数']
```

```python
all_ch[cols] = all_ch[cols].apply(pd.to_numeric, errors = 'coerce')
```

```python
# 轉換率（新增）
all_ch['转换率'] = round((all_ch['充值人数']/all_ch['新增注册用户数']), 2)
```

```python
all_ch.isna().sum()
```

```python
all_ch[all_ch['转换率'].isna()] # append連帶合併欄位，前面再改
```

```python
all_ch[all_ch['转换率'].isna()]
all_ch['转换率'] = all_ch['转换率'].fillna(0)
```

```python
na_id = all_ch[all_ch['充值人数'].isna()].index # 與新增注册用户数na.index相同
```

```python
all_ch.drop(na_id, inplace=True) 
```

```python
#inf -> 0
```

```python
all_ch[all_ch['转换率']==np.inf]
```

```python
all_ch.loc[(all_ch['转换率']==np.inf), '转换率'] = 0
```

```python
all_ch[cols]=all_ch[cols].astype(np.int32)
```

# 轉換率->float->%

```python
all_ch['转换率']=all_ch['转换率'].round(4)
f1=lambda x: '%.4f' % x
all_ch['转换率'] = all_ch['转换率'].apply(f1)
all_ch['转换率'] = all_ch['转换率'].astype('float64')
f2=lambda x: '%.2f%%' % (x*100)
all_ch['转换率'] = all_ch['转换率'].apply(f2)
```

```python
all_ch
```

```python
# 篩選需要時間
today = str(np.datetime64('2023-07-23') ) # 之後改回來today = str(np.datetime64('today'))
past_week = str(np.datetime64(today)-np.timedelta64(6,'D')) # 月報拉30天資料
filt = (all_ch['统计日期'].between(past_week, today)) # 日報需前六天
#filt
all_ch = all_ch.loc[filt]
```

```python
today 
```

```python
past_week
```

```python
# 要從7021開始做，將渠道名分割留下數字部分
def splitname(x):
    return x[0:4]
l1 = all_ch['渠道名称'].tolist()
l1 = list(map(splitname, l1))
all_ch['渠道名称']=l1
```

```python
all_ch.drop(all_ch[all_ch['渠道名称']=='代理邀请'].index, inplace=True)
all_ch.drop(all_ch[all_ch['渠道名称']=='官网渠道'].index, inplace=True)
#all_ch.drop(all_ch[all_ch['新增注册用户数']==0].index, inplace=True) 
all_ch.reset_index()
```

# 製作「潛在」成長比較分析function
做不出來 可能是 series_data_month 格式問題

```python
# 原始資料series_data_month為各系列在每個月分的獲利表
import pandas as pd
series_data_month = pd.read_csv('D:/user/anaconda workspace/HahowPython 行銷資料科學/章節 4-1 到 4-7 -【Python銷售資料處理實戰練習】/series_data_month.csv')
```

```python
def potential_growth(data, series, rank, year1, year2):
    import pandas as pd
    import plotly.express as px
    from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
    # 選擇要分析的系列
    series_data_month_growth = series_data_month[series_data_month['系列'] == series]
    # 用年份條件篩選2018, 2019年資料
    series_data_month_growth = series_data_month_growth[series_data_month_growth['訂單時間_年份'] >= int(year1.strip('年'))]

    ###
    # 測試有無前一年份可比較 條件true or false
    #test = len(series_data_month_growth[series_data_month_growth['訂單時間_年份'] == int(year2.strip('年'))]) - len(series_data_month_growth[series_data_month_growth['訂單時間_年份'] == int(year1.strip('年'))])
    #if test == 0:
    #    test == '有初始年份可以比較!'
    #    print('有初始年份可以比較!')
    #elif test > 0:
    #    test = "注意"+ str(year1) +"資料不全"
    #    print("注意"+ str(year1) +"資料不全")
    #elif test < 0:
    #    test = "注意"+ str(year2) +"資料不全"
    #    print("注意"+ str(year2) +"資料不全")
    ###
    
    test = series_data_month_growth[series_data_month_growth['訂單時間_年份']==int(year1.strip('年'))]

    if len(test) ==0:
        print('沒有初始年份可以比較')
        test = '跳過'
        series_data_month_growth = '沒有'
        return series_data_month_growth, test
    else:
        print('有初始年份可以比較!')
        test = '有資料'

    # 切出月份 series_data_month_growth['月份']，先轉成str
    series_data_month_growth['訂單時間_月份'] = series_data_month_growth['訂單時間_月份'].astype(str)
    series_data_month_growth['月份'] = series_data_month_growth['訂單時間_月份'].str.split('-', expand = True)[1]

    # 根據「訂單時間_年份」轉換成pivot
    series_data_month_growth = series_data_month_growth.pivot(index = ['月份', '系列'], columns = '訂單時間_年份', values = 'profit')
    # 格式調整
    series_data_month_growth = series_data_month_growth.reset_index()
    series_data_month_growth = series_data_month_growth.rename(columns={int(year1.strip('年')):year1, int(year2.strip('年')):year2 })
    series_data_month_growth = round(series_data_month_growth)

    # 選資料齊全來計算
    series_data_month_growth = series_data_month_growth[0:10]
    
    # 計算series_data_month_growth['年增成長']和['總年增成長率']
    series_data_month_growth['年增成長'] = round( (series_data_month_growth[year2] - series_data_month_growth[year1] )/ series_data_month_growth[year1], 2)
    series_data_month_growth['總年增成長率'] = (series_data_month_growth[year2][0:10].sum() - series_data_month_growth[year1][0:10].sum())/ series_data_month_growth[year1][0:10].sum()

    

    # 輸出所有series的2018/2019獲利成長比較表
    series_data_month_growth.to_csv( 'Top' +  str(rank) +'_高獲利系列_「' + series + '」_「潛在」成長比較分析表.csv', encoding='utf-8-sig')

    # 繪圖
    # rank = 1 
    # series = '系列3'
    # 再次繪製折線圖：
    fig = px.line(series_data_month_growth, 
                x="月份", 
                y="年增成長", 
                #   color = '訂單時間_年份',
                title ='Top' +  str(rank) +'_高獲利系列_「' + series + '」_「潛在」成長比較分析圖')

    # 如何加入背景元素（layout）
    fig.update_layout(
                    title={
                        'text':  'Top' +  str(rank) +'_高獲利系列_「' + series + '」_「潛在」成長比較分析圖',
                        'y':0.9,
                        'x':0.5,
                        'xanchor': 'center',
                        'yanchor': 'top'},
                    xaxis_title='月份',
                    yaxis_title='年增成長')
    fig.update_traces(mode="lines+markers")
    plot(fig, filename = 'Top' +  str(rank) +'_高獲利系列_「' + series + '」_「潛在」成長比較分析圖.html')

    # rank = rank +1
    return series_data_month_growth, test
```

# 進行系列「穩定成長」 分析

```python
len(series_data_month_growth_all)
```

```python
series_data_month_growth_all
```

```python
stable_df = series_data_month_growth_all.dropna()
```

```python
stable_df_tmp = stable_df[stable_df['系列'] == '系列3']
```

```python
stable_df_tmp
```

```python
# 製作「正常年增率」與「穩定指標（標準差）」
# 如果系列中沒有超過9個月的話，即跳過；若超過，則製作
# 新增['正常年增率']：去除缺值後，才具比較標準，計算年增率才公允
# 新增['穩定指標']（標準差）：計算每一個系列年增成長的標準差。若數值低，則成長效果較為穩健
```

```python
if len(stable_df_tmp) <9:
    print('跳過')

else:
    stable_df_tmp['正常年增率'] = (stable_df_tmp['2019年'].sum() - stable_df_tmp['2018年'].sum())/stable_df_tmp['2018年'].sum()

    stable_df_tmp['穩定指標（標準差）'] = stable_df_tmp['年增成長'].std()
```

```python
stable_df_tmp
```

# For迴圈製作每個系列「正常年增率」與「穩定指標（標準差）」的狀況

```python
stable_df['系列'].unique()
```

```python
stable_df_tmp_list = []
for series in stable_df['系列'].unique():
    stable_df_tmp = stable_df[stable_df['系列'] == series]
    
    if len(stable_df_tmp) <9:
        print('跳過')

    else:
        stable_df_tmp['正常年增率'] = (stable_df_tmp['2019年'].sum() - stable_df_tmp['2018年'].sum())/stable_df_tmp['2018年'].sum()
        stable_df_tmp['穩定指標（標準差）'] = stable_df_tmp['年增成長'].std()
        stable_df_tmp_list.append(stable_df_tmp)
```

```python
stable_df_final = pd.concat(stable_df_tmp_list, axis = 0)
```

```python
stable_df_final
```

```python
stable_df_final.to_csv('2_「穩定」成長比較分析.csv', encoding='utf-8-sig')
```

# 排序並生成「穩定指標值」 的分析表

```python
stable_df_index = stable_df_final[['系列', '正常年增率', '穩定指標（標準差）']]
```

```python
stable_df_index
```

```python
stable_df_index = stable_df_index.drop_duplicates()
```

```python
stable_df_index
```

```python
# 標準差由小至大排序
stable_df_index = stable_df_index.sort_values('穩定指標（標準差）', ascending = False)
```

```python
# 新增穩定排名
stable_df_index['穩定指標高低'] = range(1, len(stable_df_index)+1)
```

```python
# ['穩定指標值'] = 正常年增率	* '穩定指標高低'
stable_df_index['穩定指標值'] = stable_df_index['正常年增率'] * stable_df_index['穩定指標高低']
```

```python
# 由大至小排序
stable_df_index = stable_df_index.sort_values('穩定指標值', ascending = False)
```

```python
# 再排名
stable_df_index['建議參考順序'] = range(1, len(stable_df_index)+1)
```
